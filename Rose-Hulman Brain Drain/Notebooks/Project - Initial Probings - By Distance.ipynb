```python
def get_counts(data_Input):
    grouped_By_location = data_Input.groupby(['StateCity'])
    distance = np.zeros(len(grouped_By_location))
    area = np.zeros(len(grouped_By_location))
    Lats_ThreeD = np.zeros(len(grouped_By_location))
    Lons_ThreeD = np.zeros(len(grouped_By_location))
    Marker = np.zeros(len(grouped_By_location))
    i = 0
    
    for name, group in grouped_By_location:
        distance[i] = grouped_By_location.get_group(name).iloc[0]['DistFromRoseHulman']
        area[i] = (len(grouped_By_location.get_group(name)))
        Lats_ThreeD[i] = grouped_By_location.get_group(name).iloc[0]['lats']
        Lons_ThreeD[i] = grouped_By_location.get_group(name).iloc[0]['lons']
        i = i + 1
        d = {'Distances': pd.Series(distance, index = range(len(distance))),\
                'Repeats': pd.Series(area, index = range(len(distance))),\
                'Lats': pd.Series(Lats_ThreeD, index = range(len(distance))),\
                'Lons': pd.Series(Lons_ThreeD, index = range(len(distance)))}
    return pd.DataFrame(d)
```

```python
def draw_Locations(data, Alpha_Value):
    # create new figure, axes instances.
    fig=plt.figure()
    ax=fig.add_axes([0.1,0.1,0.8,0.8])
    
    llcrnrlat = data['lats'].min()
    urcrnrlat = data['lats'].max()
    lat_mid = (llcrnrlat + urcrnrlat)/2.0
    
    llcrnrlon = data['lons'].min()
    urcrnrlon = data['lons'].max()
    lon_mid = (llcrnrlon + urcrnrlon)/2.0
    
    # setup mercator map projection.
    m = Basemap(llcrnrlon,llcrnrlat, urcrnrlon, urcrnrlat,\
                rsphere=(6378137.00,6356752.3142),\
                resolution='h',projection='merc',lon_0=lon_mid, lat_0=lat_mid, lat_ts = 20)
#    m = Basemap(projection='ortho', lon_0=lon_mid, lat_0=lat_mid )
    
#    for i in range(len(data)):
#        # currentlat, currentlon are lat/lon of the city currently being processed.
#        currentlat = data.ix[i,6]
#        currentlon = data.ix[i,7]
#        m.drawgreatcircle(RoseHulmanLon,RoseHulmanLat,currentlon,currentlat,linewidth=.1,color='r', alpha=.7)
        
    x = data.lats
    y = data.lons      
    x1,y1=m(y,x)
    m.plot(x1,y1,'ro', alpha = Alpha_Value)
    
    m.drawmapboundary(fill_color='black')
    m.drawcoastlines()
    m.fillcontinents(color='white',lake_color='black',zorder=0)
    m.drawstates()
    m.drawcountries()
#    for xpt,ypt,name in zip(x1,y1,data.City):
#        plt.text(xpt+10000,ypt+10000,name, fontsize=8)

    ax.set_title('Where Rose goes')
    plt.show()
```
