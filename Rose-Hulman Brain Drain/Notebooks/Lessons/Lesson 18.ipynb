```python
P = prices.values
```

```python
P[0:5,0:5]
```

```python
X = sklearn.preprocessing.scale(P, axis = 1)
```

```python
X.mean(axis = 1)
```

```python
X.std(axis = 1)
```

```python
X[0,0:10]
```

```python
ward = sklearn.cluster.Ward(n_clusters=100)
```

```python
ward.fit(X)
```

```python
ward.labels_
```

```python
prices.index.values
```

```python
data = np.array([prices.index.values, ward.labels_]).T
```

```python
clusters = pd.DataFrame(data, columns = ['Ticker','Cluster'])
```
