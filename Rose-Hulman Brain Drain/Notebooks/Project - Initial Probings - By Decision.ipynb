```python
data = pd.read_csv('Newest_No_Outliers_Salary.csv', header = 0).drop(['Unnamed: 0'], axis = 1)
```

```python
data.Major.unique()
```

```python
major_Dict = {}
major_Dict['BE CS'] = 'BE','CS'   
major_Dict['ME EE'] = 'ME','EE'
major_Dict['CHE CHEM'] = 'CHE', 'CHEM'
major_Dict['CHE BE'] = 'CHE', 'BE'
major_Dict['SE CS'] = 'SE', 'CS'
major_Dict['MA ECON'] = 'MA'
major_Dict['CIVIL'] = 'CE'
```

```python
drop_indicies = []
temp_Majors = []
temp_States = []

for index, row in data.iterrows():
    if major_Dict.has_key(row['Major']):
        if major_Dict[row['Major']] == 'MA':
            temp_Majors.append('MA')
            temp_States.append(row['State'])
            
        elif major_Dict[row['Major']] == 'CE':
            temp_Majors.append('CE')
            temp_States.append(row['State'])
            
        else:
            for entry in major_Dict[row['Major']]:
                temp_Majors.append(entry)
                temp_States.append(row['State'])
        drop_indicies.append(index)
```

```python
temp_Majors
```

```python
By_state = pd.DataFrame({"Major" : temp_Majors,"State" : temp_States})
```

```python
data_dropped = data.drop(data.index[drop_indicies])
data_dropped.index.names = [None]
data_dropped.index = np.arange(len(data_dropped))
```

```python
data_fixed = data_dropped.append(By_state, ignore_index = True)
data_fixed.index.names = [None]
data_fixed.index = np.arange(len(data_fixed))
```

```python
data_fixed.Major.unique()
```

```python
data_fixed.State = data_fixed.State.apply(lambda x: 1 if x == 'IN' else 0)
```

```python
grouped_by_Major = data_fixed.groupby('Major')
```

```python
group_Count = []
group_Percentage = []
group_Major = []
for name, group in grouped_by_Major:
    group_Percentage.append(1.0*len(group[group['State'] == 'IN'])/len(group['State']))
    group_Count.append(len(group.Major))
    group_Major.append(name)
```

```python
group_Count = []
group_Percentage = []
group_STD = []
group_Major = []
for name, group in grouped_by_Major:
    group_Percentage.append(group['State'].mean())
    #if group['State'].std() np.n:
    #    print 'true'
    #print group['State'].std()
    #std = 0 if group['State'].std() == np.nan else group['State'].std()
    group_STD.append(group['State'].std())
    group_Count.append(len(group.Major))
    group_Major.append(name)
```

```python
By_Indiana = pd.DataFrame({"State" : group_Percentage, "Major" : group_Major, "Std" : group_STD, "Samples" : group_Count})
```

```python
By_Indiana.Std = By_Indiana.Std.fillna(0)
```
