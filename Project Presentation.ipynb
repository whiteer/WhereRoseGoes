```python
#Universalize input strings

def strip(inputString):
    if inputString == "" | inputString.upper() == 'Na':
        return np.NAN
    return inputString.strip().replace("-", " ")

def stripAndUpperCase(inputString):
    if inputString == "":
        return np.NAN
    return inputString.strip().replace("-", " ").upper()

def stripAndUpperCaseRemoveSlash(inputString):
    if inputString == "":
        return np.NAN
    return inputString.strip().replace("/", " ").upper()

def stripAndTitle(inputString):
    if inputString == "":
        return np.NAN
    return inputString.strip().replace("-", " ").title()
```

# Data-wrangling: Use Google API to get Lat/Lons

```python
#Gather all the unique City|State pairs to minimalize Google API requests
uniqueCityState = data.CityState.unique()
```

```python
#Gets a latitude, longitude, county tuple from Google Maps API given a city|State pair

def getGeoCoordinates((City, State)):
    
    #Create the right query given the input tuple
    replaceString = "http://maps.googleapis.com/maps/api/geocode/json?address=+REPLACEMECITY,+REPLACEMESTATE&sensor=false"    
    replaceString = replaceString.replace("REPLACEMECITY", City.replace(" ", "+"))
    replaceString = replaceString.replace("REPLACEMESTATE", State)
    
    #Query the API
    XMLstream = urllib2.urlopen(replaceString).read()
    
    #Extract the lat/lon values
    Location_Index = XMLstream.find("location")
    lat = float(XMLstream[XMLstream.find("lat\" : ",Location_Index) + 7:XMLstream.find(",",Location_Index)])
    lng = float(XMLstream[XMLstream.find("lng\" : ",Location_Index) + 7:XMLstream.find("\n",XMLstream.find("lng\" : ",Location_Index))])
    
    #Extract the county value and remove any of the trailing "County" idenifiers for standardization
    County_Index = XMLstream.rfind("short_name",0,XMLstream.find("administrative_area_level_2")) + 15
    County = XMLstream[County_Index: XMLstream.find("\n",County_Index)  - 2]
    if " County" in County:
        County = County[0: County.find(" County")]
    
    #Rest so not to upset the Google gods
    time.sleep(.5)
    
    #return a dictionary
    return {(City, State): (lat, lng, County)}
```

```python
#Create a list of dictionaries of all the lat/lon/counties given each unique city/state pair
dictionary_Lats_Longs = np.vectorize(getGeoCoordinates)(uniqueCityState)
```

```python
#Create a single dictionary from the list of dictionaries
Dictionary_Lats_Longs = {}
for i in range(dictionary_Lats_Longs.size):
    for key in dictionary_Lats_Longs[i]:
        Dictionary_Lats_Longs[key] = []
        Dictionary_Lats_Longs[key].extend(dictionary_Lats_Longs[i][key])
```

```python
#Given a city|State input thir returns the the lat/lon/county for entry into a dataframe
def assign_coordinates((City, State)):
    return Dictionary_Lats_Longs[(City, State)][0], Dictionary_Lats_Longs[(City, State)][1], Dictionary_Lats_Longs[(City, State)][2]
```

```python
#Add the lat/lon/county data to the dataframe 
data['lats'], data['lons'], data['County']  = zip(*data.CityState.map(assign_coordinates))
```

# Get Headings/Distances

```python
RoseHulmanLat = 39.4798449
RoseHulmanLon = -87.3307898
```

```python
# Calculate the great circle distance between two points on the earth (specified in decimal degrees)

def haversine(lon2, lat2):
    lat1 = RoseHulmanLat
    lon1 = RoseHulmanLon
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 

    # 6367 km is the radius of the Earth, .621371 miles in a km
    miles = 0.621371 * 6367 * c
    
    
    Bearing = atan2(cos(lat1)*sin(lat2)-sin(lat1)*cos(lat2)*cos(lon2-lon1), sin(lon2-lon1)*cos(lat2))
    return miles, degrees(Bearing)
```

```python
data['DistFromRoseHulman'], data['BearingFromRoseHulman'] = np.vectorize(haversine)(data['lons'],lat2 = data['lats'])
```

```python
def get_counts(data_Input):
    grouped_By_location = data_Input.groupby(['CityState'])
    distance = np.zeros(len(grouped_By_location))
    area = np.zeros(len(grouped_By_location))
    Lats_ThreeD = np.zeros(len(grouped_By_location))
    Lons_ThreeD = np.zeros(len(grouped_By_location))
    Marker = np.zeros(len(grouped_By_location))
    i = 0
    
    for name, group in grouped_By_location:
        distance[i] = grouped_By_location.get_group(name).iloc[0]['DistFromRoseHulman']
        area[i] = (len(grouped_By_location.get_group(name)))
        Lats_ThreeD[i] = grouped_By_location.get_group(name).iloc[0]['lats']
        Lons_ThreeD[i] = grouped_By_location.get_group(name).iloc[0]['lons']
        i = i + 1
        d = {'Distances': pd.Series(distance, index = range(len(distance))),\
                'Repeats': pd.Series(area, index = range(len(distance))),\
                'Lats': pd.Series(Lats_ThreeD, index = range(len(distance))),\
                'Lons': pd.Series(Lons_ThreeD, index = range(len(distance)))}
    return pd.DataFrame(d)
```

```python
def gen_javascript(data, div_id=None):

    if div_id == None:
        div_id = key
    # Creates Javascript objects which will comprise geoData.
    coords = ',\n  '.join(["new google.maps.LatLng(%s, %s)" % tuple(pair) for pair in data])    
    template_jscript = """
    var geoData = [
    %s
    ];
        
    var map, heatmap;
        
    function hmap_initialize() {
    
        var mapOptions = {
            zoom: 4,
            center: new google.maps.LatLng(39.8282, -98.5795),
            panControl: false,
            zoomControl: false,
            scaleControl: true,
            mapTypeControl: true,
            mapTypeControlOptions: {
              style: google.maps.MapTypeControlStyle.DROPDOWN_MENU
            }
        };
        
        map = new google.maps.Map(document.getElementById('%s'), mapOptions);
        
        var styleArray = [
            {
                featureType: "road.highway",
                elementType: "labels.text.stroke",
                stylers: [ 
                    { visibility: "off" } 
                ] 
            }
        ];

        map.setOptions({styles: styleArray});
        
        var pointArray = new google.maps.MVCArray(geoData);
        
        heatmap = new google.maps.visualization.HeatmapLayer({ data: pointArray});
        
        heatmap.setMap(map);
    }
   
    hmap_initialize();
    
    var gradient = [
        'rgba(0, 0, 0, 0)',
        'rgba(255, 0, 0, 1)',
        'rgba(128, 0, 85, 1)',    
        'rgba(0, 0, 170, 1)',
        'rgba(0, 0, 255, 1)',
        'rgba(0, 128, 255, 1)',
        'rgba(0, 255, 255, 1)'
    ]
      
    heatmap.set('gradient', gradient);
    heatmap.set('radius', 20);
    heatmap.set('opacity', 0.7);
"""
    return template_jscript % (coords, div_id)
```

```python
for county_id in us_counties:
    if us_counties[county_id]['state'] in ['ak', 'hi', 'pr', 'gu', 'vi', 'mp', 'as']:
        continue
    if County_Count_Dict.has_key(us_counties[county_id]['name']):
        if County_State_Dict[us_counties[county_id]['name']] == us_counties[county_id]['state'].upper():
            rate = County_Count_Dict[us_counties[county_id]['name']]
            idx = min(int(rate/6), 5) + 1
            county_colors.append(colors[idx])
        else:
            county_colors.append(colors[0])
    else:
        county_colors.append(colors[0])
```
